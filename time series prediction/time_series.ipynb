```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
from datetime import datetime, date
from statsmodels.tsa.api import ExponentialSmoothing, Holt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt
```

```python
Nestle = pd.read_csv('nestle.csv')
Nestle.head()
Nestle.tail()
```

```python
Nestle1 = Nestle[['Date','close']]
Nestle1.head()
```

```python
Nestle1.plot(figsize=(12,10), linewidth=4, fontsize=15)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.title('Price_Movement',fontsize=20)
```

```python
#Creating train and test set 
Nestle_train=Nestle1[0:150] 
Nestle_test=Nestle1[150:]
```

```python
y_hat_avg = Nestle_test.copy()

fit1 = Holt(np.asarray(Nestle_train['close'])).fit(smoothing_level = 0.2,smoothing_slope = 0)
y_hat_avg['Holt_linear'] = fit1.forecast(len(Nestle_test))

plt.figure(figsize=(16,8))
plt.plot(Nestle_train['close'], label='Train')
plt.plot(Nestle_test['close'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(Nestle_test.close, y_hat_avg.Holt_linear))
print(rms)
```

```python
y_hat_avg = Nestle_test.copy()
fit1 = ExponentialSmoothing(np.asarray(Nestle_train['close']) ,seasonal_periods=52 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(Nestle_test))
plt.figure(figsize=(16,8))
plt.plot(Nestle_train['close'], label='Train')
plt.plot(Nestle_test['close'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(Nestle_test.close, y_hat_avg.Holt_Winter))
print(rms)
```
