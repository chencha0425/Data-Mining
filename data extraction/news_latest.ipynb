```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import requests 
```

```python
#chrome=r"C:\chencha\chromedriver.exe"
#driver = webdriver.Chrome(executable_path=chrome)
driver = webdriver.Chrome('C:/chencha/chromedriver.exe')
```

```python
url='https://www.klsescreener.com/v2/'
driver.get(url)
```

```python
#Click the 'Screen' button on the page
screen_button=driver.find_element_by_id('submit')
screen_button.click()
```

```python
html=driver.page_source
soup=BeautifulSoup(html)
```

```python
table=soup.find(id="result")
```

```python
stock_code = []
for row in table:
    odd_row=table.find_all(class_='list odd')
    even_row=table.find_all(class_='list even')
    for code in odd_row:
        stockcode =code.find('td',attrs={'title':'Code'}).get_text()
        stock_code.append(stockcode)
    for item in even_row:
        stockcode =code.find('td',attrs={'title':'Code'}).get_text()
        stock_code.append(stockcode)
```

```python
stock_code
```

```python
stock_code_unique=[]
for i in stock_code: 
    if i not in stock_code_unique: 
        stock_code_unique.append(i) 
```

```python
mylinks = list(dict.fromkeys(stock_code_unique))
mylinks
```

```python
company_links = []
for n in stock_code_unique:
    link = 'https://www.klsescreener.com/v2/news/stock/' + str(n)
    company_links.append(link)
company_links
```

```python
for l in company_links:
    print(l)
```

```python
News=[]
News_Date=[]
company_link=[]
for link in company_links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    news_table=soup.find('div',id="section")
    for news in news_table.find_all(class_="figcaption"):
        News.append(news.get_text())
    for date in news_table.find_all(class_="item-title-secondary subtitle"):
        News_Date.append(date.get_text())
        company_link.append(link)
```

```python
company_link
```

```python
df=pd.DataFrame({'code':company_link, 'date':News_Date,'news':News})
```

```python
df
```

```python
df['company_code']=df['code'].str.split('stock/').str.get(1)
df
```

```python
df.to_csv("news_latest.csv",index=False,sep=',')
```
