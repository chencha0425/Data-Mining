```python
import time
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('C:/chencha/chromedriver.exe')
```

```python
pages=[]
for i in range (1,4):
    url='http://www.bursamalaysia.com/market/listed-companies/company-announcements/#/?category=all' +str(i)
    pages.append(url)
```

```python
for item in pages:
    driver.get(item)
    time.sleep(5)
    html=driver.page_source
    soup=BeautifulSoup(html)
```

```python
table=soup.find('div', id="bm_ajax_container")
Ann_date=[]
Comp_name=[]
Title=[]
links=[]
Comp_link=[]
Ann_link=[]
```

```python
for row in table.find_all('tr'):
    data_cells=row.find_all('td')
    data_cells2=row.find_all('a')
    for link in row.find_all('a'):
        data_cells3=link.get('href')
        links.append(data_cells3)
        #links is an array with even number element=company link, odd number element=announcement link
        Comp_link=links[::2]
        Ann_link=links[1::2]
    if len(data_cells)==4:
        Ann_date.append(data_cells[1].get_text())
        Comp_name.append(data_cells2[0].get_text())
        Title.append(data_cells2[1].get_text()) 
```

```python
Stock_announcement=pd.DataFrame({'Announcement_Date':Ann_date, 'Company_Name':Comp_name, 
                                 'Title':Title, 'Company_link':Comp_link, 'Announcement_link':Ann_link})
Stock_announcement
```

```python
#to split the company link to get the company code, after split it will become a string with two columns
#left column=company link, right column=company code
Stock_announcement['Company_Code']=Stock_announcement['Company_link'].str.split('=').str.get(1)
Stock_announcement
```
